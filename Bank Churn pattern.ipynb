```python
df = df.drop(['CustomerId', 'Surname'], axis=1)
```

```python
# Verified successfully all rows and columns 
# Now create the Segmentations 
df
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 30, 45, 60, 100],
    labels=['<30', '30-45', '46-60', '60+'])
```

```python
df['CreditScore_value']=pd.qcut(df['CreditScore'],q=3,labels=['Low','Medium','High'])
```

```python
df['Balance_Value']=pd.cut(df['Balance'],bins=[-1,0,df['Balance'].median(),df['Balance'].max()],
                                               labels=['Zero-Balance','Low-Balance','High-Balance'])
```

```python
df['Tenure_term']=pd.qcut(df['Tenure'],q=3,labels=['New','Mid-Term','Long-Term'])
```

```python
df['Financial_Score'] = (
    df['Balance'].rank(pct=True) +
    df['EstimatedSalary'].rank(pct=True) +
    df['CreditScore'].rank(pct=True) +
    df['NumOfProducts'].rank(pct=True)
).round(2)

df['Financial_Profile'] = df['Financial_Score'].apply(
    lambda x: 'High Value' if x >= df['Financial_Score'].median() else 'Low Value'
)
```
